```python
import glob
path_list = glob.glob("C:\\Users\\egork\\mlschl\\Seminars\\20_newsgroups\\sci.space\\*")
```

```python
from collections import Counter

def printDict(d):
    answer = sorted(dict(d).items(), key = lambda item: -item[1])
    for key, value in answer:
        print(key, " - ", value)

for path in path_list:
    print("File with path", path, ": ")
    cnt = Counter()
    file = open(path)
    inp = file.read().split()
    for word in inp:
        cnt[word] += 1
    printDict(cnt)
```
